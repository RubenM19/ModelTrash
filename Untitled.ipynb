```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('./dataset/data_2.csv')
```

```python
df.head()
```

```python
plt.style.use('fivethirtyeight')
df['OUTPUT'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
sns.pairplot(data=df,hue='OUTPUT')
plt.show()
```

```python
df['OUTPUT'].replace(['Normal','Abnormal'],[1,0],inplace=True)
```

```python
df['OUTPUT'].unique()
```

```python
x = df.drop('OUTPUT',axis=1)
y = df['OUTPUT']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred_train = model.predict(x_train)
```

```python
test = confusion_matrix(y_pred,y_test)
```

```python
plt.title('Test Data')
sns.heatmap(data=test,annot=True,fmt='g')
plt.show()
```

```python
train = confusion_matrix(y_pred_train,y_train)
```

```python
plt.title('Train Data')
sns.heatmap(data=train,annot=True,fmt='g')
plt.show()
```

```python
model2 = SVC(C=5,gamma='auto')
model2.fit(x_train,y_train)
```

```python
print('Test Accuracy:',model2.score(x_test,y_test))
print('Train Accuracy:',model2.score(x_train,y_train))
```

```python
y_pred_train2 = model2.predict(x_train)
y_pred2 = model2.predict(x_test)
```

```python
test2 = confusion_matrix(y_pred2,y_test)
```

```python
plt.title('Test Data')
sns.heatmap(data=test2,annot=True,fmt='g')
plt.show()
```
